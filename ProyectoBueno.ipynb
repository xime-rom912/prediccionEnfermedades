```python
#Datos
import copy
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import tensorflow as tf 
import keras 
from keras.models import Sequential 
from keras.layers import Dense, Dropout 
from keras.callbacks import EarlyStopping 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import confusion_matrix, roc_auc_score, roc_curve 
from time import time 
import itertools
import pickle

#DOWNLOAD_ROOT = "https://archive.ics.uci.edu/ml/machine-learning-databases/hepatitis/hepatitis.data"
df = pd.read_csv("/content/sample_data/new_model.csv")
```

```python
df.head()
```

```python
df.replace('?',np.nan, inplace = True)
df.isnull().sum()
```

```python
df = df.drop(['Su', 'Rbc'],axis=1)
df
```

```python
y = df['Class']
X = df.drop(['Class'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X),columns=X.columns)
```

```python
# Classification on survived 'class' (target attribute) print(y) # Split dataset in training and test datasets 
entrenamientoX,pruebasX,entrenamientoY, pruebasY = train_test_split(X, y, test_size=0.2, random_state=30, shuffle=True)
```

```python
#Definición del modelo 
model = Sequential() 
model.add(Dense(units=300, activation='relu', input_dim=11)) 
model.add(Dense(units=150, activation='relu')) 
model.add(Dense(units=75, activation='relu')) 
model.add(Dense(units=1, activation='relu')) 

#compilación del modelo 
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy']) 

#Resumen del modelo 
model.summary()
```

```python
#entrenamiento del modelo 
model.fit(entrenamientoX, entrenamientoY, validation_split=0.2, epochs=120, batch_size=1000)
```

```python
#Realiza prediccion en conjunto pruebas 
score = model.evaluate(pruebasX, pruebasY) 

#Obtiene el accuracy del modelo final 
print("Valor de accuracy en conjunto de datos de test: ",score[1])
```
